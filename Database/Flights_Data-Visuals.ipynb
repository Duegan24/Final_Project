```python
# Importing dependencies
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Cut the per distance into the distance ranges.
distance = jan_19_clean["DISTANCE"]
# distance_bins = [0, 100, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000]
# distance_bins = [0, 1000,2000,3000, 4000, 5000]
distance_bins = [0, 250, 500, 750, 1000, 1500, 2000, 2500, 5000]
distance.groupby(pd.cut(distance, distance_bins)).count()
```

```python
# Establish the distance bins and group names.
# distance_bins = [0, 100, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000]
# group_names = ["<100", "100-500", "500-1000", "1000-1500", "1500-2000", "2000-2500", "2500-3000", "3000-3500", "3500-4000", "4000-4500", "4500-5000"]
# distance_bins = [0, 1000,2000,3000, 4000, 5000]
distance_bins = [0, 250, 500, 750, 1000, 1500, 2000, 2500, 5000]
# group_names = ["<1000", "1000-2000", "2000-3000", "3000-4000", "4000-5000"]
group_names = ["<250", "250-500", "500-750", "750-1000", "1000-1500", "1500-2000", "2000-2500", "2500-5000"]
```

```python
# Categorize distance based on the bins.
jan_19_clean["Distance Range (in Miles)"] = pd.cut(distance, distance_bins, labels=group_names)

jan_19_clean.head()
```

```python
# Establishing percentage of delay per distance range
df_distance=pd.DataFrame(jan_19_clean.groupby(["Distance Range (in Miles)","DEP_DEL15"])["ORIGIN_AIRPORT_ID"].count().unstack("DEP_DEL15").reset_index().values,columns=["Distance","Not_Delayed","Delayed"])
df_distance["PercentOfDelayed"]=df_distance.Delayed /(df_distance.Delayed+ df_distance.Not_Delayed)
df_distance.sort_values(by="PercentOfDelayed",ascending=False,inplace=True)
df_distance.head()
```

```python
x_axis = df_distance["Distance"]
y_axis = df_distance["PercentOfDelayed"]
explode_values = (0.2, 0, 0, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```
