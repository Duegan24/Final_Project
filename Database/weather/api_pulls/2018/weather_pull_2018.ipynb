```python
#Import requests library and API key
import requests
from config import weather_api_key

# Import Dependencies
import pandas as pd

# Import Historical API Package
from wwo_hist import retrieve_hist_data

# Import glob dependncies
import os, glob
```

```python
# Import CSV
city_code_api_df = pd.read_csv('../../../Flight_data_files/post_hardcode_cities.csv',index_col='Unnamed: 0')
city_code_api_df
```

```python
cities_unique = []
for city in city_code_api_df['City']:
    if city not in cities_unique:
        cities_unique.append(city)
print(cities_unique)
```

```python
start_date = '01-JAN-2018'
end_date = '31-JAN-2018'
location_list = cities_unique
frequency = 1
```

```python
hist_weather_data_2018 = retrieve_hist_data(weather_api_key,
                                   location_list,
                                   start_date,
                                   end_date,
                                   frequency,
                                   location_label = False,
                                   export_csv = True,
                                   store_df = True)
```

```python
#Create variables to hold all csv files and CSVs
all_files = glob.glob(os.path.join("*.csv"))
all_df = []

# Loop through all files, create df's and add all dataframes to the dataframes list
for x in all_files:
    df = pd.read_csv(x, sep=',')
    df['file'] = x.split('/')[-1]
    all_df.append(df)
```

```python
#Merge all dataframes
merged_2018_df = pd.concat(all_df, ignore_index=True, sort=True)
merged_2018_df
```

```python
weather_2018_df = pd.DataFrame(merged_2018_df, columns = ['location','date_time','precipMM','visibility','cloudcover','windspeedKmph','humidity',])
weather_2018_df
```

```python
#Export to CSV
weather_2018_df.to_csv("../../weather_data/weather_2018_df.csv", index=False)
```
