```python
# Import Dependencies
from path import Path
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Import testing dataset
initial_2019 = pd.read_csv('Database/Data/jan_19_clean_data.csv')
initial_2020 = pd.read_csv('Database/Data/jan_20_clean_data.csv')
initial_2019
```

```python
#Create a new dataframe with only relavent testing data
machine_model_2019_df = initial_2019.filter(["DAY_OF_MONTH","OP_CARRIER_AIRLINE_ID","OP_CARRIER_FL_NUM","ORIGIN_AIRPORT_ID","ORIGIN_AIRPORT_SEQ_ID","DEST_AIRPORT_SEQ_ID","DEP_TIME","DEP_DEL15","ARR_TIME","ARR_DEL15","CANCELLED","DIVERTED","DISTANCE"])
machine_model_2019_df
```

```python
# Create Feature and Target Variables
y = machine_model_2019_df["DEP_DEL15"]
X = machine_model_2019_df.drop(columns="DEP_DEL15")
```

```python
# Create X & Y train and test variables
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=34, 
                                                    stratify=y)
X_train.shape
```

```python
# Create the Logistic Regression Model
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=34)
```

```python
# Fit the model
classifier.fit(X_train, y_train)
```

```python
# Make predictions 
y_pred = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
# Show the accuracy score of the machine learning model 
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
